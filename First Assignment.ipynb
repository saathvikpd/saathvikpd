```python
#Importing the necessary packages

import pandas as pd
import numpy as np
```

```python
#The function below reads the information in a gct file and stores it in a pandas dataframe

def read_gct(path, summary=False):
    f = open(path)
    l = []
    for line in f:
        l.append(line)
    l = l[1:]
    #data0 = pd.read_csv(path)
    shape = len(l) #data0.shape
    inds = []
    for i in np.arange(2, shape):
        inds.append(l[i].split("\t")[0])

    real_data = pd.DataFrame(np.zeros((shape-2, len(l[1].split("\t"))-2)), columns = l[1].split("\t")[2:], index = inds)

    for i in np.arange(2, shape):
        info = l[i].split("\t")[2:]
#        try:
        info1 = list(float(info0) for info0 in info)
        real_data.iloc[i-2,:] = info1
#         except:
#             info1 = list(float(info0) for info0 in info[1:])
#             #info1 = np.ravel([info[0], info1])
#             real_data.iloc[i-2,1:] = info1
#             real_data.iloc[i-2,0] = info[0]

    if summary:
        print(real_data.shape)
    
    return real_data
```

```python
#Paths to the gct files

path0 = "./BRCA_minimal_60x19.gct"
path1 = "./BRCA_large_20783x40.gct"
```

```python
#Function reads the first gct file and, since summary is true, it spits out the size of the dataset

data0 = read_gct(path0, summary = True)
```

```python
#Function reads the second gct file. Since summary is not mentioned, it assumes the default value of "false"

data1 = read_gct(path1)
```

```python
#Importing the package necessary for visual illustrations

import matplotlib.pyplot as plt
```

```python
#Columns in the dataset that are of concern

cols = ["A7-A0DB-normal", "A7-A13E-normal", "BH-A0B3-primary", "BH-A0B5-primary"]
```

```python
#Graphs the above columns from the second dataset as histograms

i = 0
for c in cols:
    plt.figure(i+100)
    plt.hist(data1[c], bins = 20)
    plt.xlabel(c)
    i+=1
```

```python
#A new dataframe is created where the statistical summaries of each of the samples in the second dataset are stored

data_sum = pd.DataFrame(np.zeros((data1.shape[0], 4)), columns = ["Sample name", "Mean", "Median", "Standard deviation"])
data_sum.iloc[:,0] = data1.index
for i in range(len(data_sum)):
    data_sum.iloc[i,1] = np.mean(data1.iloc[i,:])
    data_sum.iloc[i,2] = np.median(data1.iloc[i,:])
    data_sum.iloc[i,3] = np.std(data1.iloc[i,:])
```

```python
#Summarized dataset

data_sum
```

```python
#The new dataframe is sorted based on the standard deviation column

data_sorted = data_sum.sort_values(by = ["Standard deviation"], ignore_index = True)
```

```python
#Sorted dataset

data_sorted
```

```python
#We want only the top 90% of the data, the data with low standard deviation values, so we cutoff the bottom 10%

cutoff = 9*(len(data_sorted)//10)

data_sorted = data_sorted.iloc[:cutoff, :]
```

```python
#Sorted dataset with only 90% of the data

data_sorted
```

```python
#Names of the samples that made the cut

samples = data_sorted["Sample name"]
```

```python
#The names of the samples are used to filter the original dataset

data2 = data1.loc[samples, :]
```

```python
#Another dataframe is created with the new mean, median, and standard deviation values

data3 = pd.DataFrame(np.zeros((data2.shape[0], 4)), columns = ["Sample name", "Mean", "Median", "Standard deviation"])
data3.iloc[:,0] = data2.index
for i in range(len(data3)):
    data3.iloc[i,1] = np.mean(data2.iloc[i,:])
    data3.iloc[i,2] = np.median(data2.iloc[i,:])
    data3.iloc[i,3] = np.std(data2.iloc[i,:])
```

```python
#Dataset with new statistical summaries

data3
```
